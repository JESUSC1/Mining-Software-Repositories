```python
"""
This code processes multiple CSV files containing commit data for software repositories. 
It allows the user to select which files to process, creates two plots for each selected file: 
one for the number of commits by year and another for the number of deletions, insertions, and net lines by year. 
The plots are saved as separate PDF files with a filename derived from the repository name. 
The code uses the Pandas library to read and manipulate the data and the Matplotlib library to create the plots. 
It also uses the os module to navigate file paths and the PdfPages module to save the plots as PDF files. 
The final output is a message confirming that processing is complete.
"""

import os
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

# ask the user to select files from the directory
csv_dir = "/Users/jesuscantu/Desktop/Workspace/MiningSoftwareRepositories/Raw_Data/Commits" # "path/to/your/csv/directory"
csv_files = os.listdir(csv_dir)
print("Available CSV files:")
for i, f in enumerate(csv_files):
    print(f"{i+1}. {f}")
selected_files = input("Enter the number(s) of the file(s) you want to process (separated by commas): ")
selected_files = [int(i) for i in selected_files.split(",")]

for file_idx in selected_files:
    # read in the CSV file
    df = pd.read_csv(os.path.join(csv_dir, csv_files[file_idx-1]))

    # convert the "author_date" column to a datetime object in UTC timezone
    df['author_date'] = pd.to_datetime(df['author_date'], utc=True)

    # create a new column for the year
    df['Year'] = df['author_date'].dt.year

    # group the data by year, and count the number of commits, deletions, insertions, and net lines in each group
    commits_by_year = df.groupby(['Year']).size().reset_index(name='Commits')
    deletes_by_year = df.groupby(['Year'])['num_deletes'].sum().reset_index(name='Deletes')
    inserts_by_year = df.groupby(['Year'])['num_inserts'].sum().reset_index(name='Inserts')
    net_lines_by_year = df.groupby(['Year'])['net_lines'].sum().reset_index(name='Net Lines')

    # create the first plot for the number of commits
    fig, ax1 = plt.subplots()
    ax1.plot(commits_by_year['Year'], commits_by_year['Commits'])
    ax1.set_xlabel('Year')
    ax1.set_ylabel('Number of Commits')
    ax1.set_title(f"Commits by Year for \nRepository: {csv_files[file_idx-1].split('_commit_data.csv')[0]}")

    # save the first plot as a PDF file
    save_dir = "/Users/jesuscantu/Desktop/Workspace/MiningSoftwareRepositories/Generated_Plots/Stage2" # "path/to/save/directory"
    filename = f"{csv_files[file_idx-1].split('_commit_data.csv')[0]}"
    with PdfPages(os.path.join(save_dir, f"{filename}_commits_by_year.pdf")) as pdf:
        pdf.savefig(fig)

    # create the second plot for the number of deletions, insertions, and net lines
    fig, ax2 = plt.subplots()
    ax2.plot(deletes_by_year['Year'], deletes_by_year['Deletes'], color='red', label='Deletions')
    ax2.plot(inserts_by_year['Year'], inserts_by_year['Inserts'], color='green', label='Insertions')
    ax2.plot(net_lines_by_year['Year'], net_lines_by_year['Net Lines'], color='blue', label='Net Lines')
    ax2.set_xlabel('Year')
    ax2.set_ylabel('Lines of Code Deleted/Inserted')
    ax2.set_title(f"Deletions, Insertions, and Net Lines by Year for \nRepository: {csv_files[file_idx-1].split('_commit_data.csv')[0]}")
    ax2.legend()

    # adjust spacing between title and plot
    fig.suptitle('', y = 1.05)

    # save the second plot as a PDF file
    with PdfPages(os.path.join(save_dir, f"{filename}_code_changes_by_year.pdf")) as pdf:
        pdf.savefig(fig)

    # display message when processing is complete
    print(f"Processing of {csv_files[file_idx-1]} is complete.\n")
```
