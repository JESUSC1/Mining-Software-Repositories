```python
"""
Part 1:
This code reads in a CSV file and extracts the year from the date column, groups the data by year,
and creates a plot with two y-axes. The plot shows the total code size and average repo size by year.
The code then adds labels and a title to the plot and displays it.
"""

import pandas as pd
import matplotlib.pyplot as plt

# Read in the CSV file

#filename = 'file.csv'
#df = pd.read_csv(filename)

df = code_size_df

# Convert the date column to datetime objects
df['date'] = pd.to_datetime(df['date'], utc=True)

# Extract the year from the datetime objects
df['year'] = df['date'].dt.year

# Group by year and calculate the total code size and repo size for each year
grouped = df.groupby('year').agg({'code_size_LOC': 'sum', 'repo_size_LOC': 'mean'})

# Create the plot with two y-axes
fig, ax1 = plt.subplots()

# Set the left y-axis for code size
ax1.set_xlabel('Year')
ax1.set_ylabel('Commit Code Size (LOC)')
ax1.plot(grouped.index, grouped['code_size_LOC'], color='blue')

# Set the right y-axis for repo size
ax2 = ax1.twinx()
ax2.set_ylabel('Mean Repo Size (LOC)')
ax2.plot(grouped.index, grouped['repo_size_LOC'], color='green')

# Add labels and a title
plt.title('Code Size and Repo Size over Time')

# Show the plot
plt.show()
```

```python
"""
Part 1: 
This code prompts the user for a directory containing CSV files, lists the files with the .csv extension, 
and asks the user which files they want to process. For each selected file, 
the code reads the data, converts the date column to datetime objects, extracts the year from the date,
groups the data by year, and creates a plot with two y-axes. 
The plot shows the total code size and average repo size by year. 
The code saves the plot as a PDF file in a specified output directory with a filename 
that includes the repository name. The code prints the progress and the output file location for each file processed.
"""

import os
import pandas as pd
import matplotlib.pyplot as plt

# Ask user for the path to the directory containing the CSV files
# directory = input("Enter the path to the directory containing the CSV files: ")

# Or hardcode it 
directory = "/Users/jesuscantu/Desktop/Workspace/MiningSoftwareRepositories/Raw_Data/Commits/Code_Size"

# List all the CSV files in the directory
files = [f for f in os.listdir(directory) if f.endswith('.csv')]

# Print the list of CSV files
print("The following CSV files were found in the directory:")
for i, f in enumerate(files):
    print(f"{i+1}. {f}")

# Ask the user which files they want to process
choices = input("Enter the numbers of the files you want to process (separated by commas): ")
choices = [int(c.strip()) for c in choices.split(',')]

# Process the selected files
for i, choice in enumerate(choices):
    # Load the CSV file
    filename = os.path.join(directory, files[choice-1])
    repo_name = files[choice-1].replace('-code_size_data.csv', '')

    print(f"Processing file {i+1}/{len(choices)}: {filename}...")

    df = pd.read_csv(filename)

    # Convert the date column to datetime objects
    df['date'] = pd.to_datetime(df['date'], utc=True)

    # Extract the year from the datetime objects
    df['year'] = df['date'].dt.year

    # Group by year and calculate the total code size and repo size for each year
    grouped = df.groupby('year').agg({'code_size_LOC': 'sum', 'repo_size_LOC': 'mean'})

    # Create the plot with two y-axes
    fig, ax1 = plt.subplots()

    # Set the left y-axis for code size
    ax1.set_xlabel('Year')
    ax1.set_ylabel('Commit Code Size (LOC)')
    ax1.plot(grouped.index, grouped['code_size_LOC'], color='blue')

    # Set the right y-axis for repo size
    ax2 = ax1.twinx()
    ax2.set_ylabel('Mean Repo Size (LOC)')
    ax2.plot(grouped.index, grouped['repo_size_LOC'], color='green')

    # Add labels and a title
    plt.title(f'Code Size and Repo Size over Time \nRepository: {repo_name}')
    
    # Add legend
    ax1.legend(['Code Size'], loc='upper left')
    ax2.legend(['Repo Size'], loc='upper right')

    # Save the plot as a PDF file in the output directory
    # output_dir = input("Enter the path to the output directory: ")
    
    # Or hardcode it
    output_dir = "/Users/jesuscantu/Desktop/Workspace/MiningSoftwareRepositories/Generated_Plots/Stage3"
    output_file = os.path.join(output_dir, f"{repo_name}_code_repo_size_by_year.pdf")
    plt.savefig(output_file)
    plt.show()

    # Close the plot
    plt.close()

    print(f"Finished processing file {i+1}/{len(choices)}: {filename}.\nPDF saved to {output_file}")
```

```python
"""
Part 2:
This code uses Dask to compute the issue density over time for a software repository. 
The code reads in two CSV files containing code size and issue data, creates a year column in each dataset, 
and calculates the issue density by dividing the number of issues created each year by the maximum code size in that year.
The code then plots the issue density over time using Matplotlib and saves the resulting plot as a PDF file, within a specific outupt directory.
The resulting plot shows the trend of issue density over time for the software repository.
"""

import os
import dask.dataframe as dd
import pandas as pd
import matplotlib.pyplot as plt

# Read in the code size and issue datasets
code_size = dd.read_csv('/Users/jesuscantu/Desktop/Workspace/MiningSoftwareRepositories/Raw_Data/Commits/Code_Size/scikit-learn-code_size_data.csv',
                        usecols=['date', 'code_size_LOC']) # path to code size data 
issues = dd.read_csv('/Users/jesuscantu/Desktop/Workspace/MiningSoftwareRepositories/Raw_Data/Issues/scikit-learn_issues.csv',
                     usecols=['created_at']) # path to issues data 
# Set Repo Name
repo_name = "scikit-learn"

# Create year column in each dataset
code_size['year'] = code_size['date'].str[:4]
issues['year'] = issues['created_at'].str[:4]

# Calculate issue density by year
issue_density = issues.groupby('year').size().compute() / code_size.groupby('year')['code_size_LOC'].max().compute()

# Plot the issue density over time
fig, ax = plt.subplots()
ax.plot(issue_density.index, issue_density.values)

# Check if the data spans more than 10 years
if len(issue_density.index) > 10:
    # Only show every second year on the x-axis
    plt.xticks(range(0, len(issue_density.index), 2), issue_density.index[::2])
else:
    plt.xticks(issue_density.index)

ax.set_title(f'Issue Density Over Time for \nRepository: {repo_name}')
ax.set_xlabel('Year')
ax.set_ylabel('Issue Density')

# Save plot 
output_dir = "/Users/jesuscantu/Desktop/Workspace/MiningSoftwareRepositories/Generated_Plots/Stage4"
output_file = os.path.join(output_dir, f"{repo_name}_mean_issue_density.pdf")

plt.savefig(output_file)
plt.show()

# Close the plot
plt.close()
```
